```python
import spacy
import re

unwanted_words = ['cleaned','and','washed','as','needed','thinly','sliced','Thin','add','adjustable','adjust','just','unripe', 'unskimmed', 
                  'more', 'more if','rings', 'I', 'have','used', 'I like adding','mashed','you', 'can',"minced",'preferably','per','about',
                  'warm','very', 'vertically','well','long','tsp','tbsp', 'uncooked','unpeeled','your','th','the','of','powdered',
                  'could', 'also', 'use','give', 'it','is', 'prefered','if','bunch','a','small', 'piece','pieces', 'pinch',"medium sized",
                  'additional','assorted','approximately','optional','broken','bowl','finely','chopped','small','cut','roasted',
                  'thick','cubes','julienned','deseeded', "de-veined",'slices', "only", "need",'taste', 'required','boiled',"big", "large",
                  'coarsely','pounded','grated','blanched','a','few','pureed','made','cubed','steamed','quartered','into','preferable',
                  'peeled','slivered','or','whisked','nicely','on','diced',' but recommended','according','overnight',"peel",
                  'soaked','in','roughly','Wet','Ingredients','Everest brand', 'Enough','any','other','s ', 'teaspoon', "one","two",
                  'tablespoons','tablespoon']
    
# Load the English language model from spaCy
nlp = spacy.load("en_core_web_sm")

# Function to clean and retain only nouns (ingredients)
def clean_and_filter_nouns(ingredient_text):
    # Remove quantities and units
    cleaned_text = re.sub(r'\d+(?:-\d+)?(?:/\d+)?\s*(?:cup|cups?|tablespoon|tablespoons?|teaspoon|teaspoons?|gms?|gram|grams?|kg|ml|inch|pound|pounds?|liters?)?\s*', ',', ingredient_text, flags=re.IGNORECASE)
    cleaned_text = re.sub(r'\d+\s+\b\w+\b', ',', cleaned_text, flags=re.IGNORECASE)##a number followed by unit
    # Add a comma before the words "to" or "for" (case-insensitive)
    cleaned_text = re.sub(r'\b(to|for|with|without|salt)\b', r', \1', cleaned_text, flags=re.IGNORECASE)
    
    # Remove strings that start with "to" or "for" (case-insensitive)
    #cleaned_text = re.sub(r',?\s*\b(to|for)\b.*?(,|$)', '', cleaned_text, flags=re.IGNORECASE)
    
    # Remove unwanted words/phrases from the ingredient text
    for word in unwanted_words:
        cleaned_text = re.sub(r'\b' + re.escape(word) + r'\b', '', cleaned_text, flags=re.IGNORECASE)
    
    # Remove extra spaces and clean up commas
    
    cleaned_text = re.sub(r'\s*,\s*', ',', cleaned_text)  # clean up commas
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()  # clean extra spaces
    cleaned_text = re.sub(r',+', ',', cleaned_text)  # remove consecutive commas
    cleaned_text = cleaned_text.strip(',')  # remove leading/trailing commas
    cleaned_text = re.sub(r'[/\.\-:*)(&"%/+]', '', cleaned_text)# Remove //, ./, -, /, :, *, +    
    cleaned_text = re.sub(r'\(\s*\)', '', cleaned_text)# Remove empty parentheses or parentheses with spaces 
    
     # Split the cleaned text into individual ingredients (assuming comma-separated)
    ingredient_list = [ingredient.strip().lower() for ingredient in cleaned_text.split(',') if ingredient.strip() 
                       #and re.sub(r'\([^\)]*$', '', ingredient) # Removes only '(' if it's not closed
                       #and re.sub(r'^[^\(]*\)', '', ingredient) # Removes only ')' if it's not opened
                       and not ingredient.strip().lower().startswith('for') 
                       and not ingredient.strip().lower().startswith('to') 
                       and not ingredient.strip().lower().startswith('with')
                      and not ingredient.strip().lower().startswith('without')]
    
    filtered_ingredients = []
    for ingredient in ingredient_list:
        # Create a spaCy document object for each ingredient
        doc = nlp(ingredient)
        # Keep only noun phrases or single nouns as ingredients
        filtered_tokens = [chunk.text for chunk in doc.noun_chunks]
        filtered_ingredients.extend(filtered_tokens)  
    return  filtered_ingredients

# Apply to your DataFrame
nohin_recipe_df["cleaned_ingredients"] = nohin_recipe_df["ingredients"].apply(clean_and_filter_nouns)

nohin_recipe_df["cleaned_ingredients"]
```

```python
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer


nltk.download('stopwords')# Download stopwords list (if not already downloaded)
nltk.download('wordnet')  # Required for lemmatization
nltk.download('omw-1.4')  # Optional: for additional language support in WordNet

# Initialize Porter Stemmer and stop words
#stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Preprocess the ingredient text: stemming and stop word removal
def preprocess_ingredients(ingredient_list):
    # Stem and remove stop words
    #return ' '.join([stemmer.stem(word) for word in ingredient_list if word not in stop_words])
    return ' '.join([lemmatizer.lemmatize(word) for word in ingredient_list if word not in stop_words])
    
# Apply preprocessing to the 'cleaned_ingredients' column (which is a list of ingredients)
recom_df["ingredient_str"] = recom_df["cleaned_ingredients"].apply(lambda x: preprocess_ingredients(x))

# Initialize the TF-IDF vectorizer, now without stop words
tfidf_vectorizer = TfidfVectorizer()

# Fit and transform the preprocessed ingredient strings into TF-IDF vectors
ingredient_vectors = tfidf_vectorizer.fit_transform(recom_df["ingredient_str"])

# Compute cosine similarity between the ingredient vectors
cosine_sim_matrix = cosine_similarity(ingredient_vectors)

# Output the cosine similarity matrix
print(cosine_sim_matrix)

# Example: similarity between the first and second recipes
similarity_between_first_and_second = cosine_sim_matrix[0][1]
print(f"Similarity between first and second recipe: {similarity_between_first_and_second}")
```
