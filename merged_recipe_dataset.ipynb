```python
import re

# Unwanted words list
unwanted_words = ['cleaned','and','washed','as','needed','thinly','sliced','Thin','add','adjustable','adjust','just',
                  'more', 'more if','rings', 'I', 'have','used', 'I like adding', 'about','mashed','you', 'can',
                  'warm','very', 'vertically','well','long','tsp','tbsp', 'uncooked','unpeeled','your','th','the','of',
                  'could', 'also', 'use','give', 'it','is', 'prefered','if','bunch','a','small', 'piece','pieces', 'pinch',
                  'additional','assorted','approximately','optional','broken','bowl','finely','chopped','small','cut',
                  'thick','cubes','julienned','deseeded', "de-veined",'slices', "only",'per', "need",'taste', 'required',
                  'coarsely','pounded','grated','blanched','a','few','pureed','made','cubed','steamed','quartered','into',
                  'peeled','slivered','or','whisked','nicely','with','on','diced',' but recommended','according',
                  'soaked','in','roughly','Wet','Ingredients','Everest brand', 'Enough','any','other','s ', 'teaspoon', 
                  'tablespoons']
    

# Function to clean the text
def clean_ingredients(ingredient_text):
    # Remove quantities and units (cups, tablespoons, etc.)
    cleaned_text = re.sub(r'\d+(?:-\d+)?(?:/\d+)?\s*(?:cup|cups?|tablespoon|tablespoons?|teaspoon|teaspoons?|gms?|gram|grams?|kg|ml|inch|pound|pounds?|liters?)?\s*', ',', ingredient_text, flags=re.IGNORECASE)
    cleaned_text = re.sub(r'\d+\s+\b\w+\b', ',', cleaned_text, flags=re.IGNORECASE)##a number followed by unit
    
    # Add a comma before the words "to" or "for" (case-insensitive)
    cleaned_text = re.sub(r'\b(to|for)\b', r', \1', cleaned_text, flags=re.IGNORECASE)
    # Remove unwanted words/phrases from the ingredient text
    for word in unwanted_words:
        cleaned_text = re.sub(r'\b' + re.escape(word) + r'\b', '', cleaned_text, flags=re.IGNORECASE)
    
    # Remove extra spaces and clean up commas
    
    cleaned_text = re.sub(r'\s*,\s*', ',', cleaned_text)  # clean up commas
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text).strip()  # clean extra spaces
    cleaned_text = re.sub(r',+', ',', cleaned_text)  # remove consecutive commas
    cleaned_text = cleaned_text.strip(',')  # remove leading/trailing commas
    cleaned_text = re.sub(r'[/\.\-:*&)(%/+]', '', cleaned_text)# Remove //, ./, -, /, :, *, +    
    cleaned_text = re.sub(r'\(\s*\)', '', cleaned_text)# Remove empty parentheses or parentheses with spaces  
    # Split the cleaned text into individual ingredients (assuming comma-separated)
    ingredient_list = [ingredient.strip() for ingredient in cleaned_text.split(',') if ingredient.strip() and not ingredient.strip().lower().startswith('for') and not ingredient.strip().lower().startswith('to')]
    return ingredient_list


recipe_df["cleaned_ingredients"] = recipe_df["ingredients"].apply(clean_ingredients)
recipe_df["cleaned_ingredients"]

# Create a unique ingredient set by flattening the list and removing duplicates
unique_recipe_ingredient_list = set(ingredient for sublist in recipe_df['cleaned_ingredients'] for ingredient in sublist if ingredient)

#Convert the set to a list
unique_recipe_ingredient_list = sorted(unique_recipe_ingredient_list)
unique_recipe_ingredient_list
```
