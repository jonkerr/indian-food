```python
import pandas as pd
import re
```

```python
cuisine_df = pd.read_csv("data/Indian_Food_Cuisine.csv")
cuisine_df
```

```python
cuisine_df.rename (columns = {"name_of_Dish":"name", "Diet_Type":"diet", "Course_name":"course", 
                                 "Discrption_of_Dish":"description", "Cuisine_name":"cuisine", 
                                 "Ingredients_of_Dish":"ingredients","Total_time":"prep_time", 
                              "Recipe_Instructions":"instructions"},inplace= True)
cuisine_df
```

```python
cuisine_df.drop(columns= ["Ratings_of_Dish","Prepration_time","Makes","Cooking_time"], inplace =True)
cuisine_df
```

```python
cuisine_df["instructions"] = cuisine_df["instructions"].apply(lambda x: re.sub(r"[\[\]']", '',x)).str.strip()
cuisine_df["instructions"]
```

```python
cuisine_df["name"].value_counts()
```

```python
cuisine_df['cuisine'] = cuisine_df['cuisine'].apply(lambda x: re.sub(r"[\[\]']", '',x))
cuisine_df['cuisine'] = cuisine_df['cuisine'].str.replace('Cuisine:', '').str.strip()
cuisine_df['cuisine'] = cuisine_df['cuisine'].str.replace('Recipes', '').str.strip()
cuisine_df['cuisine']
```

```python
cuisine_df['cuisine'].unique()
```

```python
cuisine_df["diet"]= cuisine_df["diet"].str.replace("Diet:","").str.strip()
cuisine_df["diet"]
```

```python
cuisine_df["course"]= cuisine_df["course"].str.replace("Course:","").str.strip()
cuisine_df["course"]
```

```python
cuisine_df["prep_time"]= cuisine_df["prep_time"].str.replace("M","").str.strip()
cuisine_df["prep_time"] = cuisine_df["prep_time"].astype('float')
cuisine_df["prep_time"]
```

```python
cuisine_df["Similar_Dishes"] = cuisine_df["Similar_Dishes"].apply(lambda x: re.sub(r"[\[\]']", '',x)).str.strip()
cuisine_df["Similar_Dishes"]
```

```python
cuisine_df["ingredients"][0]
```

```python
cuisine_df["ingredients"] = cuisine_df["ingredients"].apply(lambda x: re.sub(r"[\[\]']", '',x))                                                
cuisine_df["ingredients"]
```

```python
cuisine_df["ingredients"].iloc[7285]
```

```python
def remove_title_part(text):
    return re.sub(r'title=".*?">', '', text)
cuisine_df['ingredients'] = cuisine_df['ingredients'].apply(remove_title_part)
cuisine_df["ingredients"].iloc[7285]
```

```python
cuisine_df["ingredients"].value_counts()
```

```python
cuisine_df["ingredients"].iloc[4367]
```

```python
# Function to check if a string contains Hindi characters
def contains_hindi(text):
    if isinstance(text, str):  
        return bool(re.search(r'[\u0900-\u097F]', text))
    return False

# Apply the function across all columns and find rows with Hindi words
cuisine_hindi_rows = cuisine_df.applymap(contains_hindi).any(axis=1)

# Extract rows that contain Hindi words in any column
cuisine_df_with_hindi = cuisine_df[cuisine_hindi_rows]

print(f"Number of rows with Hindi words: {len(cuisine_df_with_hindi)}")

cuisine_df_with_hindi
```

```python
# Create a new DataFrame without the rows containing Hindi words
nohindi_cuisine_df = cuisine_df[~cuisine_hindi_rows]

print(f"Number of rows with Hindi words removed: {len(cuisine_df) - len(nohindi_cuisine_df)}")
```
